```python
import pandas as pd
```

```python
# Path to CSV files
csv_files = ['preProcessingAnies.csv', 'preProcessingPrabowo.csv', 'preProcessingGanjar.csv', 'preProcessingMuhaimin.csv', 'preProcessingMahfud.csv']


gibran = pd.read_csv('preProcessingGibran.csv', delimiter=';')
# Read each CSV file into a pandas DataFrame
dataframes = [pd.read_csv(csv) for csv in csv_files]

# Concatenate the DataFrames vertically (row-wise)
combined_dataframe = pd.concat(dataframes + [gibran], ignore_index=True)

# Write the combined DataFrame to a new CSV file
combined_dataframe.to_csv('dataset-pemilu.csv', index=False)
```
