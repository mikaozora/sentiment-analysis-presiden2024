# Data Cleaning

```python
df = pd.read_csv('Ganjar Pranowo.csv', encoding='latin1')
df = df[['Date', 'Tweet']]
df
```

```python
df.duplicated().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
def cleanTwitter(text):
  text = re.sub(r'@[A-Za-z0-9_]+', ' ', text)
  text = re.sub(r'#\w+', ' ', text)
  text = re.sub(r'RT[\s]+', ' ', text)
  text = re.sub(r'https?://\S+', ' ', text)
  text = re.sub(r'[^A-Za-z0-9 ]', ' ', text)
  text = re.sub(r'\s+', ' ', text).strip()

  return text

df['Tweet'] = df['Tweet'].apply(cleanTwitter)
df['Tweet'] = df['Tweet'].str.lower()
```

```python
df = df.drop_duplicates(subset=['Tweet'])
```

```python
#tokenize
tokenized = df['Tweet'].apply(lambda x:x.split())
tokenized
```

```python
#stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in text:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  return d_clean

tokenized = tokenized.apply(stemming)
tokenized.to_csv("preProcessingGanjar.csv", index=False)
```
