# Data Cleaning

```python
df = pd.read_csv('/content/drive/MyDrive/DATA_ML_kel11/data_gibran.csv')
df = df[['full_text', 'username', 'created_at']]
df
```

```python
df.duplicated().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
def cleanTwitter(text):
  text = re.sub(r'@[A-Za-z0-9_]+', ' ', text)
  text = re.sub(r'#\w+', ' ', text)
  text = re.sub(r'RT[\s]+', ' ', text)
  text = re.sub(r'https?://\S+', ' ', text)
  text = re.sub(r'[^A-Za-z0-9 ]', ' ', text)
  text = re.sub(r'\s+', ' ', text).strip()

  return text

df['full_text'] = df['full_text'].apply(cleanTwitter)
df['full_text'] = df['full_text'].str.lower()
```

```python
df = df.drop_duplicates(subset=['full_text'])
```

```python
#tokenize
tokenized = df['full_text'].apply(lambda x:x.split())
tokenized
```

```python
#stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in text:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  return d_clean

tokenized = tokenized.apply(stemming)
tokenized.to_csv("preProcessingGibran.csv", index=False)
```
